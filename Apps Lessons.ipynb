```python
dating = apps.loc[apps['Category']=='DATING']
dating = dating.reset_index()
dating1 = dating
for index,row in dating1.iterrows():
    if '$' in row['Price']:
        dating1.at[index,'Price'] = (dating1.at[index,'Price'].replace('$',''))
    if '+' in row['Installs'] or ',' in row['Installs']:
        dating1.at[index,'Installs'] = dating1.at[index,'Installs'].replace('+','') 
        dating1.at[index,'Installs'] = dating1.at[index,'Installs'].replace(',','') 
dating1['Price'] = dating1['Price'].astype(float)
dating1['Installs'] = dating1['Installs'].astype(float)
print(dating1.head())
print(dating1.dtypes)
```

```python
#make histograms to check what are the most downloaded apps. you use the seaborn library. sns.distplot, make a countplot
#find the apps that have the most installs
#make new columns with these target audiences: 

#cols = ['App']
#keywords = ['Muslim','BBW','Live','Black', 'BLK','Christian','50','Gay','Lesbian','Trans','Latin','White','Sugar','Elite', 'Rich','Herpes', 'Blind','Stranger','Rando','Random']
```

```python
keywords = ['Muslim','BBW','Live','Black', 'BLK','Christian','50','Gay','Lesbian','Trans','Latin','White','Sugar','Elite', 'Rich','Herpes', 'Blind','Stranger','Rando','Random']
dating1['Category2'] = ''
for index,row in dating1.iterrows():
    for i in keywords:
        if str(i) in row['App']:
            dating1.at[index,'Category2'] = str(i)
            
print(dating1.head())
print(dating1.dtypes)
dating2 = dating1.groupby(by='Category2').agg({'Installs':sum})
dating2.sort_values('Installs', ascending = False)
dating3 = dating1.groupby(by = 'Category2').size()
print(dating3.sort_values(ascending = False))
```
